```python
measurements = sorted(measurements)#sorting the list from smallest to largest using sorted method so that it would be easier to do the manipulations with data
def mean(measurements):
    return round(sum(measurements)/len(measurements), 2) #finding mean by adding all the measurements using the sum and dividing by the number of the elements of the list that is found using len function. rounding the answer to 2 integers using the round function

def my_range(measurements):
    return round(measurements[-1] - measurements[0],2) #range is found by substracting the lowes number(fisrt in the sorted list) from the highes(last in the sorted list)

def median(measurements):
    middle = round(len(measurements)/2) #defie the middle point of the list
    if len(measurements) % 2 == 0: #if the number o fvariables is even(can be devided by 2 without a remainder)
        median = round(((measurements[middle] + measurements[middle - 1])/2), 2) #than the median will be the average between 2 central numbers(middle and the one before it) rounded to 2 decimal
    else:
        median = measurements[middle] #is not even(odd) than the middle point is the median
    return median

def mode(measurements):
    occurences = 0 #set occurences to 0
    for i in measurements: #create a loop that will check each number
        if measurements.count(i) > occurences: #count the occurences of each number and compares with the variable occurences, if it is greater
            occurences = measurements.count(i) #so the occurences is updated with the new number
            mode = [i] #when all the numbers are cheacked the mode is the one that has the highest number of occurences
        elif measurements.count(i) == occurences: #if there are more than 1 mode, 2 items will have the same occurences
            if i not in mode: #check if the item is in the list of modes(that means that this is a different item)
                mode.append(i) #add a new item to the list
    return mode

def std_dev(measurements):
    sqr_of_mes = []  #create a list for square roots of the list
    for i in measurements: #create a loop to make that list of squates
        sqr_of_mes.append(abs(mean(measurements) - i) ** 2)#sudstact from mean each value of the list and make it squared. abs function used to aviod negative values
    varience = sum(sqr_of_mes)/(len(measurements)) #find the variance by deviding the sum of all the numbers in a list of the squares by the number of mesurements in the list
    return round(varience ** 0.5, 2)#find the standart deviation be finding the squate root of variance and roundin gthe result to 2 numbers after comma

#print all the results
print("The mean is",  mean(measurements),  "the range is", my_range(measurements), "the median is", median(measurements) ,  "the mode is", mode(measurements),  "the standart deviation is", std_dev(measurements) )   
```

### 4.1 Matching Napkins Simulation

What is the probability that you are unlucky, and sadly NONE of the napkins are matched to the correct measurement (you guessed all of them wrong)? **Write a simulation in Python to estimate this probability**. Ensure that your code has sufficiently detailed comments to explain the key steps in the solution. **Note**: *Use your own data for this simulation*. So, if there are repeated values in your measurements, take that into consideration: what if two napkins for two distinct measurements show the same value, and are then swapped? Would those napkins be considered a mismatch, or a lucky match? It’s up to you to decide how to run the simulation as long as you clearly explain your approach. 

Hints to get started (use these if you wish, keeping in mind that there are multiple approaches):

- You can define your stack of napkins with measurements in the correct order as a list in Python. For example, you can set `napkins = [measurement1,measurement2,...]`. This means that napkins[0] should give the first measurement. 
- A random permutation of this list can be created with the following code: `rand_napkins = np.random.choice(napkins,N,replace=False)`. You should be able to explain how this function works, including the parameters passed, and why it is relevant for the problem. Read the [Numpy documentation](https://numpy.org/doc/stable/reference/random/generated/numpy.random.choice.html) to learn more.
- If you have $N$ measurements, you will want to check whether `rand_napkins[i] == napkins[i]`, for each value of `i` from $0$ to $N-1$.
- You’ll need to use a loop to create many random lists and repeat the checking procedure, keeping track of the number of matches each time.
- More Python tips at the end of the assignment instructions in Forum.


```python
#import libraries needed for work
import numpy as np
import random

lst_match = [] #list for storing the history of matches in simulation
n= 10000 #to increase the reliability set 10,000 as a number the simulation will be done
for n in range(n): #create a loop for running a simulation
    rand_mesure = np.random.choice(measurements,10,replace=False) #generate a list of randomly replaced napkins from the original list
    match = 0 #set the starting value of match as 0 inside the first loop but ouside the second, so that it was renewed before each comapison
    for i in range(0, len(measurements)): #create a loop to check with range corresponding the numer of elements in the list
        if rand_mesure[i] == measurements[i]: #check if the valuses in both list with the same positionl index are the same
            match +=1 #is they correspond, consider it as match and add 1 to all the marches of this list
    lst_match.append(match) #add the total numer of matched napkins from this round of simulation to the list
result_0 = lst_match.count(0)/n#calculate the number of apperances of 0 in the list of all matches and devide by the numer of times the simulation was run
result_1 = lst_match.count(1)/n #appearances of 1
result_2 = lst_match.count(2)/n #appearances of 2
result_3 = lst_match.count(3)/n #appearances of 3

print('The probability that NONE of the napkins are matched to the correct measurement is {} or {}%, \nthe probability of 1 match is {} or {}%, \nthe probability of 2 matches is {} or {}%, \nthe probability of 3 matches is {} or {}%'.format(result_0, round(result_0*100), result_1, round(result_1*100), result_2,round(result_2*100), result_3, round(result_3*100))) #print the result outputing the data in 2 ways: decimal number and persentades(round used for rounding)

```
